# Matplotlib for Beginners: Visual Features Explained

This notebook explains how Matplotlib works and how to use its most common chart features.

You will learn:
- What `plt` means
- How a figure and axes work
- How to create line, scatter, bar, and histogram charts
- How to add titles, labels, legends, grids, colors, annotations, and subplots
- How to save a chart as an image


## 1. Import Libraries

Matplotlib is a Python visualization library. Most notebooks import `matplotlib.pyplot` as `plt`. The `plt` name is short, common, and easy to type.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("Matplotlib is ready")
```

## 2. Create Simple Practice Data

We will use a small sales dataset so every chart is easy to understand.


```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [120, 150, 170, 160, 210, 240]
profit = [25, 35, 45, 38, 60, 72]
customers = [80, 95, 110, 105, 140, 165]

business_data = pd.DataFrame({
    "month": months,
    "sales": sales,
    "profit": profit,
    "customers": customers,
})

business_data
```

## 3. Understand Figure and Axes

A **figure** is the full drawing area. An **axes** is the actual chart inside that area.

`plt.subplots()` gives you both:
- `figure`: the full canvas
- `axis`: the chart area where data is drawn


```python
figure, axis = plt.subplots(figsize=(8, 5))

axis.plot(business_data["month"], business_data["sales"])
axis.set_title("Monthly Sales")
axis.set_xlabel("Month")
axis.set_ylabel("Sales")

plt.show()
```

## 4. Add a Line Chart

Use a line chart when you want to show change over time. The line connects values in order.


```python
plt.figure(figsize=(8, 5))
plt.plot(business_data["month"], business_data["sales"], marker="o")
plt.title("Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

## 5. Add a Scatter Chart

Use a scatter chart when you want to compare two numeric values. Each dot is one row of data.


```python
plt.figure(figsize=(8, 5))
plt.scatter(business_data["customers"], business_data["sales"], s=100)
plt.title("Sales by Customer Count")
plt.xlabel("Customers")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

## 6. Add a Bar Chart

Use a bar chart when you want to compare categories. Here each month is one category.


```python
plt.figure(figsize=(8, 5))
plt.bar(business_data["month"], business_data["profit"])
plt.title("Profit by Month")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.grid(axis="y")
plt.show()
```

## 7. Add a Histogram

Use a histogram to understand the spread of values. It groups numbers into ranges called bins.


```python
np.random.seed(42)
daily_sales = np.random.normal(loc=180, scale=35, size=100)

plt.figure(figsize=(8, 5))
plt.hist(daily_sales, bins=12)
plt.title("Distribution of Daily Sales")
plt.xlabel("Sales Range")
plt.ylabel("Number of Days")
plt.grid(axis="y")
plt.show()
```

## 8. Add Colors, Line Styles, and Legends

A legend explains what each line or color means. This is important when one chart has more than one series.


```python
plt.figure(figsize=(8, 5))
plt.plot(business_data["month"], business_data["sales"], marker="o", color="blue", label="Sales")
plt.plot(business_data["month"], business_data["profit"], marker="s", color="green", linestyle="--", label="Profit")
plt.title("Sales and Profit")
plt.xlabel("Month")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```

## 9. Add Text Annotations

An annotation points to an important value. Use it to highlight a peak, drop, or business event.


```python
best_month_index = business_data["sales"].idxmax()
best_month = business_data.loc[best_month_index, "month"]
best_sales = business_data.loc[best_month_index, "sales"]

plt.figure(figsize=(8, 5))
plt.plot(business_data["month"], business_data["sales"], marker="o")
plt.title("Sales with Best Month Highlighted")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.annotate(
    "Best month",
    xy=(best_month, best_sales),
    xytext=(best_month_index - 1, best_sales + 20),
    arrowprops={"arrowstyle": "->"},
)
plt.grid(True)
plt.show()
```

## 10. Add Multiple Charts with Subplots

Subplots let you show multiple related charts in one figure. This is useful for dashboards and comparisons.


```python
figure, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].plot(business_data["month"], business_data["sales"], marker="o")
axes[0].set_title("Sales Trend")
axes[0].set_xlabel("Month")
axes[0].set_ylabel("Sales")
axes[0].grid(True)

axes[1].bar(business_data["month"], business_data["profit"])
axes[1].set_title("Profit by Month")
axes[1].set_xlabel("Month")
axes[1].set_ylabel("Profit")
axes[1].grid(axis="y")

plt.tight_layout()
plt.show()
```

## 11. Save a Chart

`savefig()` saves the current chart to an image file. In real projects, this is useful for reports, presentations, and documentation.


```python
plt.figure(figsize=(8, 5))
plt.plot(business_data["month"], business_data["sales"], marker="o")
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.savefig("monthly_sales_chart.png", dpi=150, bbox_inches="tight")
plt.show()
```

## 12. Quick Matplotlib Feature Map

| Feature | Code Example | Purpose |
| --- | --- | --- |
| Figure size | `plt.figure(figsize=(8, 5))` | Controls chart size |
| Line chart | `plt.plot(x, y)` | Shows trend over time |
| Scatter chart | `plt.scatter(x, y)` | Shows relationship between values |
| Bar chart | `plt.bar(x, y)` | Compares categories |
| Histogram | `plt.hist(values)` | Shows value distribution |
| Title | `plt.title("Title")` | Names the chart |
| Axis labels | `plt.xlabel()`, `plt.ylabel()` | Explains axes |
| Legend | `plt.legend()` | Explains series names |
| Grid | `plt.grid(True)` | Improves readability |
| Annotation | `plt.annotate()` | Highlights a point |
| Subplots | `plt.subplots()` | Creates multiple charts |
| Save image | `plt.savefig()` | Exports the chart |

Beginner rule: start with the simplest chart, then add title, labels, grid, and legend only when needed.
